## ***Import All the Required Libraries***

```python
# Import Required Libraries
from langchain.document_loaders import UnstructuredURLLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma
from langchain.chains import RetrievalQA
from transformers import TFAutoModelForCausalLM, AutoTokenizer, pipeline
from langchain import HuggingFacePipeline
import sys
```

## ***Pass the URLs and extract the data from these URLs***

```python
# Pass URLs and Extract Data
URLs = [
    'https://blog.gopenai.com/paper-review-llama-2-open-foundation-and-fine-tuned-chat-models-23e539522acb',
    'https://www.mosaicml.com/blog/mpt-7b',
    'https://stability.ai/blog/stability-ai-launches-the-first-of-its-stablelm-suite-of-language-models',
    'https://lmsys.org/blog/2023-03-30-vicuna/'
]
loaders = UnstructuredURLLoader(urls=URLs)
data = loaders.load()
```

## ***Split the Text into Chunks***

```python
# Split Text into Chunks
text_splitter = CharacterTextSplitter(separator='\n', chunk_size=1000, chunk_overlap=200)
text_chunks = text_splitter.split_documents(data)
```

## ***Download the Hugging Face Embeddings***

```python
# Download Hugging Face Embeddings (No API Key Required)
embeddings = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
```

## ***Convert the Text Chunks into Embeddings and Create a Knowledge Base***

```python
# Convert Text Chunks into Embeddings and Create Knowledge Base in ChromaDB
vectorstore = Chroma.from_documents(text_chunks, embeddings, persist_directory="./chroma_db")
```

## ***Create a Large Language Model (LLM) Wrapper***

```python
# Load a Small Model Without Quantization
model_name = "distilgpt2"  # Lightweight model
```

```python
# Load Tokenizer and TensorFlow Model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = TFAutoModelForCausalLM.from_pretrained(model_name)
```

```python
# Create a Text Generation Pipeline
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, framework="tf", max_new_tokens=256, do_sample=True)
```

```python
# Wrap Pipeline with LangChain
llm = HuggingFacePipeline(pipeline=pipe)
```

## ***Initialize the Retrieval QA with Source Chain***

```python
# Initialize the Retrieval QA Chain
qa = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=vectorstore.as_retriever())
```

```python
# Example Queries
queries = [
    "How good is Vicuna?",
    "How does Llama 2 outperform other models?",
    "What is StableLM?"
]
for query in queries:
    print(f"Query: {query}")
    result = qa.run(query)
    print(f"Answer: {result}\n")
```

```python
# Interactive Chatbot
while True:
    user_input = input("Input Prompt: ")
    if user_input.lower() == 'exit':
        print('Exiting...')
        sys.exit()
    if user_input.strip() == '':
        continue
    result = qa.run(user_input)
    print(f"Answer: {result}\n")
```
